```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("dstyle")
_ = plt.figure()
plt.style.use("dstyle")
```

```python
table1_fname = "integrands/integrandONE_vals.txt"
table2_fname = "integrands/integrandTWO_vals.txt"
table3_fname = "integrands/integrandTHREE_vals.txt"
table4_fname = "integrands/integrandFOUR_vals.txt"
table5_fname = "integrands/integrandFIVE_vals.txt"
table6_fname = "integrands/integrandSIX_vals.txt"
table7_fname = "integrands/integrandSEVEN_vals.txt"

table1 = np.loadtxt(table1_fname, skiprows=6, delimiter = ',')
table2 = np.loadtxt(table2_fname, skiprows=6, delimiter = ',')
table3 = np.loadtxt(table3_fname, skiprows=6, delimiter = ',')
table4 = np.loadtxt(table4_fname, skiprows=6, delimiter = ',')
table5 = np.loadtxt(table5_fname, skiprows=6, delimiter = ',')
table6 = np.loadtxt(table6_fname, skiprows=6, delimiter = ',')
table7 = np.loadtxt(table7_fname, skiprows=6, delimiter = ',')
```

## Test 1

$$
\int_{a=0}^x \frac{4}{1+x^2} \textrm{d}x = 4 (\arctan(x) - \arctan(a=0)) = 4 \arctan(x)
$$

```python
x_int = table1[:,0]

int_val = table1[:,1]
analytical_val = 4.*np.arctan(x_int)

abs_err = np.abs(int_val - analytical_val)

frac_err = abs_err/analytical_val
```

```python
fig, ax = plt.subplots(figsize=(20,5), nrows=1, ncols=3)

ax[0].scatter(x_int, int_val, label='Me')
ax[0].plot(x_int, analytical_val, label='Numpy')
ax[0].legend()

ax[1].plot(x_int, abs_err)
ax[1].set_yscale('log')

ax[2].plot(x_int, frac_err)
ax[2].set_yscale('log')
```

## Test 2

$$
\int_{a=0}^x \cos(x) \textrm{d}x = \sin(x) - \sin(0) = \sin(x)
$$

```python
x_int = table2[:,0]

int_val = table2[:,1]
analytical_val = np.sin(x_int)

abs_err = np.abs(int_val - analytical_val)

frac_err = abs_err/analytical_val
```

```python
fig, ax = plt.subplots(figsize=(20,5), nrows=1, ncols=3)

ax[0].scatter(x_int, int_val, label='Me')
ax[0].plot(x_int, analytical_val, label='Numpy')
ax[0].legend()

ax[1].plot(x_int, abs_err)
ax[1].set_yscale('log')

ax[2].plot(x_int, frac_err)
ax[2].set_yscale('log')
```

## Test 3

$$
\int_{a=0}^x \frac{1}{\alpha x + \beta} \textrm{d}x = \frac{\ln{ |\alpha x + \beta| } } { \alpha } - \frac{\ln{ |\alpha 0 + \beta| } } { \alpha } = \frac{\ln{ |\alpha x + \beta| } } { \alpha } - \frac{\ln{ |\beta| } } { \alpha } = \frac{1}{\alpha} \ln \left(\frac{\alpha}{\beta} x + 1\right)
$$

```python
alpha, beta = 0.25, 2./3.
x_int = table3[:,0]

int_val = table3[:,1]
analytical_val = (1./alpha)*np.log(((alpha/beta) * x_int) + 1)

abs_err = np.abs(int_val - analytical_val)

frac_err = abs_err/analytical_val
```

```python
fig, ax = plt.subplots(figsize=(20,5), nrows=1, ncols=3)

ax[0].scatter(x_int, int_val, label='Me')
ax[0].plot(x_int, analytical_val, label='Numpy')
ax[0].legend()

ax[1].plot(x_int, abs_err)
ax[1].set_yscale('log')

ax[2].plot(x_int, frac_err)
ax[2].set_yscale('log')
```

## Test 4

$$
\int_{a=0}^x \frac{1}{(1+z)\sqrt{\alpha(1+z)^3}} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \int_{a=0}^x (1+z)^{-5/2} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \frac{-2}{3} \left( \left( 1 + x  \right)^{-3/2} - 1 \right)
$$

$$
\int_{a=0}^x \frac{1}{\sqrt{\alpha(1+z)^3}} \textrm{d}z = \frac{2}{3\sqrt{\alpha}} \left(1 - \left(1 + x \right)^{-3/2} \right)
$$

```python
alpha = 0.25
lnx_int = table4[:,0]
x_int = np.exp(lnx_int)

int_val = table4[:,1]
analytical_val = (2./(3.*np.sqrt(alpha)))*(1. - ((1. + x_int)**(-3/2)))

abs_err = np.abs(int_val - analytical_val)

frac_err = abs_err/analytical_val
```

```python
fig, ax = plt.subplots(figsize=(20,5), nrows=1, ncols=3)

ax[0].scatter(x_int, int_val, label='Me')
ax[0].plot(x_int, analytical_val, label='Numpy')
ax[0].set_xscale('log')
ax[0].legend()

ax[1].plot(x_int, abs_err)
ax[1].set_yscale('log')
ax[1].set_xscale('log')

ax[2].plot(x_int, frac_err)
ax[2].set_yscale('log')
ax[2].set_xscale('log')
```

## Test 5

$$
\int_{a=0}^x \frac{1}{(1+z)\sqrt{\alpha(1+z)^4}} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \int_{a=0}^x (1+z)^{-3} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \frac{-1}{2} \left( \left( 1 + x  \right)^{-2} - 1 \right)
$$

$$
\int_{a=0}^x \frac{1}{\sqrt{\alpha(1+z)^4}} \textrm{d}z = \frac{1}{2\sqrt{\alpha}} \left(1 - \left(1 + x \right)^{-2} \right)
$$

```python
alpha = 0.25
lnx_int = table5[:,0]
x_int = np.exp(lnx_int)

int_val = table5[:,1]
analytical_val = (1./(2. * np.sqrt(alpha)))*(1. - ((1. + x_int)**(-2)))

abs_err = np.abs(int_val - analytical_val)

frac_err = abs_err/analytical_val
```

```python
fig, ax = plt.subplots(figsize=(20,5), nrows=1, ncols=3)

ax[0].scatter(x_int, int_val, label='Me')
ax[0].plot(x_int, analytical_val, label='Numpy')
ax[0].set_xscale('log')
ax[0].legend()

ax[1].plot(x_int, abs_err)
ax[1].set_yscale('log')
ax[1].set_xscale('log')

ax[2].plot(x_int, frac_err)
ax[2].set_yscale('log')
ax[2].set_xscale('log')
```

## Test 6

$$
\int_{a=0}^x \frac{1}{(1+z)\sqrt{\alpha(1+z)^2}} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \int_{a=0}^x (1+z)^{-2} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \frac{-1}{1} \left( \left( 1 + x  \right)^{-1} - 1 \right) = \frac{1}{\sqrt{\alpha}} \left(\frac{1 + x}{1 + x} - \frac{1}{1 + x} \right)
$$

$$
\int_{a=0}^x \frac{1}{(1+z)\sqrt{\alpha(1+z)^2}} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \left(\frac{x}{1 + x} \right)
$$

```python
alpha = 0.25
lnx_int = table6[:,0]
x_int = np.exp(lnx_int)

int_val = table6[:,1]
analytical_val = (1./np.sqrt(alpha))*(x_int / (1 + x_int))

abs_err = np.abs(int_val - analytical_val)

frac_err = abs_err/analytical_val
```

```python
fig, ax = plt.subplots(figsize=(20,5), nrows=1, ncols=3)

ax[0].scatter(x_int, int_val, label='Me')
ax[0].plot(x_int, analytical_val, label='Numpy')
ax[0].set_xscale('log')
ax[0].legend()

ax[1].plot(x_int, abs_err)
ax[1].set_yscale('log')
ax[1].set_xscale('log')

ax[2].plot(x_int, frac_err)
ax[2].set_yscale('log')
ax[2].set_xscale('log')
```

## Test 7

$$
\int_{a=0}^x \frac{1}{(1+z)\sqrt{\alpha(1+z)}} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \int_{a=0}^x (1+z)^{-3/2} \textrm{d}z = \frac{1}{\sqrt{\alpha}} \frac{-2}{1} \left( \left( 1 + x  \right)^{-1/2} - 1 \right)
$$

$$
\int_{a=0}^x \frac{1}{(1+z)\sqrt{\alpha(1+z)^2}} \textrm{d}z = \frac{2}{\sqrt{\alpha}} \left(1 - \left( 1 + x  \right)^{-1/2} \right)
$$

```python
fig, ax = plt.subplots(figsize=(20,5), nrows=1, ncols=3)

ax[0].scatter(x_int, int_val, label='Me')
ax[0].plot(x_int, analytical_val, label='Numpy')
ax[0].set_xscale('log')
ax[0].legend()

ax[1].plot(x_int, abs_err)
ax[1].set_yscale('log')
ax[1].set_xscale('log')

ax[2].plot(x_int, frac_err)
ax[2].set_yscale('log')
ax[2].set_xscale('log')
```
