```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("dstyle")
_ = plt.figure()
plt.style.use("dstyle")
```

```python
MPC_CGS     = 3.0857e24
KM_CGS      = 1.e5
GYRS_CGS    = 3.1536e16

H0 = 67.66

H0_cgs = H0 * (KM_CGS / MPC_CGS)
```

## Test 1

$$
\int_{z}^\infty \frac{1}{(1+z)\sqrt{\Omega_{\textrm{m}}(1+z)^3}} \textrm{d}z = \frac{1}{\sqrt{\Omega_{\textrm{m}}}} \int_{z}^\infty (1+z)^{-5/2} \textrm{d}z = \frac{1}{\sqrt{\Omega_{\textrm{m}}}} \frac{-2}{3} \left( \left( 1 + \infty  \right)^{-3/2} - \left( 1 + z  \right)^{-3/2} \right)
$$

$\Omega_{\textrm{m}} = 1.0 $

$$
\int_{z}^\infty \frac{1}{\sqrt{\Omega_{\textrm{m}}(1+z)^3}} \textrm{d}z = \frac{2}{3} \left(1 + z \right)^{-3/2}
$$

```python
table_Matter_fname = "table_matter.txt"

table = np.loadtxt(table_Matter_fname, skiprows=1, delimiter = ',')

z_arr = table[:,0]
l_z_arr = table[:,1]
a_arr = table[:,2]

time_look_arr = table[:,3]
time_age_arr = table[:,4]

hubble_arr = table[:,5]
density_arr = table[:,6]
```

```python
expected_age = (2./3.)*(1 + z_arr)**(-3/2)
expected_age_s = expected_age / H0_cgs
expected_age_Gyr = expected_age_s / GYRS_CGS
```

```python
calculated_age_s = time_age_arr / H0_cgs
calculated_age_Gyr = calculated_age_s / GYRS_CGS
```

```python
abs_err = np.abs(calculated_age_Gyr - expected_age_Gyr)
frac_err = abs_err/expected_age_Gyr
```

```python
fig, ax = plt.subplots(figsize=(15,6), nrows=1, ncols=3)

ax[0].scatter(z_arr, calculated_age_Gyr, label='Me')
ax[0].plot(z_arr, expected_age_Gyr, label='Numpy')

# ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].set_title("Age of Universe (Gyr)")
ax[0].set_xlabel('redshift')
ax[0].legend()


ax[1].plot(z_arr, abs_err)

ax[1].set_yscale('log')
# ax[1].set_xscale('log')
ax[1].set_xlabel('redshift')
ax[1].set_title("Absolute Error")

ax[2].plot(z_arr, frac_err)

ax[2].set_yscale('log')
# ax[2].set_xscale('log')
ax[2].set_xlabel('redshift')
ax[2].set_title("Fractional Error")

fig.suptitle(r"$\Omega_{\rm{m}}$ = 1")

fig.tight_layout()
```

## Test 7

$$
\int_{z}^{\infty} \frac{1}{(1+z)\sqrt{\Omega_{\Lambda}(1+z)^0}} \textrm{d}z = \frac{1}{\sqrt{\Omega_{\Lambda}}} \int_{z}^{\infty} \frac{1}{1+z} \textrm{d}z = \frac{1}{\sqrt{\Omega_{\Lambda}}} \left(\ln(\infty) - \ln(z+1)\right)
$$

$$
\int_{z}^{\infty} \frac{1}{(1+z)\sqrt{\Omega_{\Lambda}(1+z)^2}} \textrm{d}z = \ln \left( \frac{\infty}{z + 1} \right)
$$

$$
\int_{z}^{\infty} \frac{1}{(1+z)\sqrt{\Omega_{\Lambda}(1+z)^2}} \textrm{d}z = \ln \left( \frac{10^5}{z + 1} \right)
